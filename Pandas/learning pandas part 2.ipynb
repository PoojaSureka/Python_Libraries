```python
def run_scores(batsman_name):
    vk = deliveries[deliveries['batsman'] == batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).index[0]
```

```python
run_scores("V Kohli")
```

```python
run_scores("DA Warner")
```

```python
run_scores("G Gambhir")
```

```python
run_scores("MS Dhoni")
```

```python
# Strike Rate = (number of runs/ number of balls)*100
# min batsman 200 balls in over 16-20
```

```python
death_over = deliveries[deliveries['over']>15]
```

```python
all_batsman = death_over.groupby("batsman")['batsman_runs'].count()
```

```python
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
```

```python
# runs scored by all these 43 batsman
#balls played by these 43 batsman
```

```python
final = death_over[death_over["batsman"].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```
