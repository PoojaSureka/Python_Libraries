Bar chart

```python
### bivariate analysis
### numerical vs categorical
### use case - aggregate analysis of groups
```

```python
children = [10,20,30,25,40]
colors = ["red", "blue", "green", "yellow", "pink"]
```

```python
### vertical barcharts
plt.bar(colors, children, color = "green" )
```

```python
### horizontal barcharts
plt.barh(colors, children, color = "blue" )
```

```python
df = pd.read_csv("batsman_season_record.csv")
```

```python
df
```

```python
plt.bar(df["batsman"], df["2015"], width = 0.5 )
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df["2015"], width =0.2, color = "green")
plt.bar(np.arange(df.shape[0]),df["2016"], width =0.2, color = "red")
plt.bar(np.arange(df.shape[0]) + 0.2,df["2017"], width =0.2, color = "blue")

plt.xticks(np.arange(df.shape[0]), df["batsman"])
plt.show()
```

```python
children = [10,20,30,25,40]
colors = ["redredredredred", "blueblueblueblueblue", "greengreen", "yellow", "pink"]

plt.bar(colors, children, color = "green" )
```

```python
children = [10,20,30,25,40]
colors = ["redredredredred", "blueblueblueblueblue", "greengreen", "yellow", "pink"]
plt.bar(colors, children, color = "blue" )

plt.xticks(rotation = "vertical")
plt.show()
```

stacked bar chart

```python
data = [23,45,100,49,20]
subject = ["english", "science", "math", "hindi", "evs"]
plt.pie(data, labels=subject)
plt.show()
```
