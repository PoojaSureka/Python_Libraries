```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
```

```python
sns.violinplot(data = data, x= "total_bill")
```

```python
sns.violinplot(data = data, x = "day", y = "total_bill", hue = "sex")
```

```python
sns.violinplot(data = data, x = "day", y = "total_bill", hue = "sex", split = "True")
```
