```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
data = sns.load_dataset("flights")
```

```python
data.head()
```

```python
x = data.pivot_table(index = "year", columns =  "month" , values = "passengers", aggfunc= "sum")
```

```python
sns.heatmap(x)
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric="correlation")
```

```python
sns.clustermap(x, z_score=0, metric="correlation", annot = True)
```

```python
sns.clustermap(x, z_score=0, metric="correlation",row_cluster= False,  annot = True)
```

```python
sns.clustermap(x, z_score=0, metric="correlation", col_cluster= False, annot = True)
```
