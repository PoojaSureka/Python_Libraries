```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
data = sns.load_dataset("tips")

### jointplot are used to do bivariate analysis on numerical data
```

```python
data.head()
```

```python
sns.jointplot(data = data , x = "total_bill", y = "tip", kind = "scatter")
```

```python
sns.jointplot(data = data , x = "total_bill", y = "tip", kind = "hex")
```

```python
sns.jointplot(data = data , x = "total_bill", y = "tip", kind = "kde")
```

```python
sns.jointplot(data = data , x = "total_bill", y = "tip", kind = "reg")
```
