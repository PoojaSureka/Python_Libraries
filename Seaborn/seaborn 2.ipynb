```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### categorical Scatterplots
### x - axis = categorical data
### y - axis = numerical data
```

```python
plt.style.use("fivethirtyeight")
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
sns.catplot(data = data, x = "day" , y = "tip", kind = "strip" )
```

```python
sns.catplot(data = data, x = "day" , y = "tip", kind = "strip", jitter = 0 )
```

```python
sns.catplot(data = data, x = "day" , y = "tip", kind = "swarm" )
```

```python
sns.catplot(data = data, x = "day" , y = "tip", kind ="swarm", hue = "sex" )
### we do not have size and style parameter over here
```

```python
sns.swarmplot(data = data, x = "day" , y = "tip", hue = "sex" )
```
