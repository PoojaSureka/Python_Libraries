```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
data = sns.load_dataset("flights")
```

```python
data.head()
```

```python
### heatmap are basically drawn between two categorical data 
### and sometime between numerical data but it should be discrete numerical data
```

```python
data.pivot_table(index = "year", columns =  "month" , values = "passengers")
```

```python
x = data.pivot_table(index = "year", columns =  "month" , values = "passengers", aggfunc= "sum")
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False, linewidths = 1.5, annot= True, fmt = 'd')
```

```python
sns.heatmap(x, cbar=False, linewidths = 1.5, annot= True, fmt = 'd', cmap = "viridis")
```

```python
sns.heatmap(x, cbar=False, linewidths = 1.5, annot= True, fmt = 'd', cmap = "summer")
```

```python
sns.heatmap(x, cbar=False, linewidths = 1.5, annot= True, fmt = 'd', cmap = "winter")
```
