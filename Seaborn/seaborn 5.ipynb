```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
```

```python
sns.countplot(data = data , x = "sex")
```

```python
sns.catplot(data = data , x = "smoker", y = "total_bill", kind = "bar")
```
